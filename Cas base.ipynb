Quantificació i mostreig aleatori d'ADN

Cas base --> 1 sol nucleòtid

generar_cadena crea una llista amb *length* nucleòtids amb distribució uniforme (25% per A, T, C, G)

PCR_transition és 1 cicle de PCR. Genera un valor aleatori entre 0 i 1. Si el valor és més gran que la probabilitat de canvi *p*, el nucleòtid es duplica i no canvia. Si és més petit, el nucleòtid es duplica però en un altre nucleòtid. Per exemple, si la cadena té 1 nucleòtid A, la probabilitat de duplicar-se a T és p/3, a C és p/3 i a G és p/3.

```python
import random
```

```python
# Genera una cadena d'ADN aleatòria de longitud especificada

def generar_cadena(length):

    cadena = []                                                         # Incialitzar cadena buida

    for idx in range(length):                                           # Iterar per la longitud desitjada
        random_nucleotid = random.choice(['A', 'T', 'C', 'G'])          # Escollir un nucleòtid aleatori
        cadena.append(random_nucleotid)                                 # Afegir el nucleòtid a la cadena
    return cadena
```

```python
# Realitza un cicle de PCR amb probabilitat de transició p

def PCR_transition(cadena, p):

    cadena_post_PCR = []                                                        # Inicialitzar cadena buida per a la nova cadena després de PCR

    for idx in range(len(cadena)):                                              # Iterar sobre tota la cadena d'ADN
        
        random_value = random.random()                                          # Generar un valor aleatori entre 0 i 1

        if random_value > p:                                                    # El nucleòtid es duplica sense canviar amb probabilitat 1 - p
            cadena_post_PCR.append(cadena[idx])                                 # Afegir el mateix nucleòtid a la llista
        
        else:                                                                   # El nucleòtid canvia amb probabilitat p --> eg. si A, passa a T amb p/3, C amb p/3 i G amb p/3                    
            if cadena[idx] == 'A':
                cadena_post_PCR.append(random.choice(['T', 'C', 'G']))

            elif cadena[idx] == 'T':
                cadena_post_PCR.append(random.choice(['A', 'C', 'G']))

            elif cadena[idx] == 'C':
                cadena_post_PCR.append(random.choice(['A', 'T', 'G']))

            elif cadena[idx] == 'G':
                cadena_post_PCR.append(random.choice(['A', 'T', 'C']))
    
    return cadena_post_PCR                                                      # Retornar la nova cadena després de PCR

# Aquesta funció hauria de retornar una nova llista amb sa cadena duplicada (i probablement modificada) o afegir els nous nucleòtids a la llista existent?
```

```python
# Realitza c cicles de PCR

def cicles_PCR(cadena, c, p):

    for idx in range(c):
        PCR_transition(cadena, p)
```

```python
def contador(nucleotids):
    conteig = {'A': 0, 'T': 0, 'C': 0, 'G': 0}

    for idx in nucleotids:
        conteig[idx] += 1
        
    return conteig
```

```python
cadena = generar_cadena(1)

print(cadena)

contador_cadena = contador(cadena)
print(contador_cadena)



cadena_post_PCR = PCR_transition(cadena, 0.5)

print(cadena_post_PCR)

contador_resultat = contador(cadena_post_PCR)
print(contador_resultat)
```
